```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")

tabela_freq = df["Pclass"].value_counts().reset_index()
tabela_freq.columns = ["Classe", "Quantidade"]
tabela_freq["Percentual (%)"] = (tabela_freq["Quantidade"] / tabela_freq["Quantidade"].sum() * 100).round(2)

fig, ax = plt.subplots(figsize=(6, 2))
ax.axis('tight')
ax.axis('off')
tabela = ax.table(cellText=tabela_freq.values, colLabels=tabela_freq.columns, cellLoc='center', loc='center')
tabela.auto_set_font_size(False)
tabela.set_fontsize(12)
tabela.scale(1.2, 1.5)
plt.title("Tabela de Frequência - Classe dos Passageiros", fontsize=14, fontweight='bold')
plt.savefig("tabela_frequencia.png", dpi=300, bbox_inches='tight')
plt.show()

df["Sex"].value_counts().plot(kind='bar', color=['blue', 'pink'], edgecolor='black')
plt.title("Quantidade de Homens e Mulheres no Titanic")
plt.xlabel("Sexo")
plt.ylabel("Quantidade")
plt.savefig("grafico_barras.png")
plt.show()

df["Pclass"].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Distribuicao por Classe")
plt.savefig("grafico_pizza.png")
plt.show()

df["Age"].dropna().plot(kind='hist', bins=20, edgecolor='black', color='green')
plt.title("Distribuicao de Idade dos Passageiros")
plt.xlabel("Idade")
plt.ylabel("Quantidade")
plt.savefig("histograma.png")
plt.show()

df.boxplot(column=["Fare"])
plt.title("Boxplot - Tarifas Pagas")
plt.ylabel("Tarifa")
plt.savefig("boxplot.png")
plt.show()

print("\n=== VARIAVEL 1: IDADE ===")
idade = df["Age"].dropna()
print(f"Media: {idade.mean():.2f}")
print(f"Mediana: {idade.median():.2f}")
print(f"Moda: {idade.mode()[0]:.2f}")
print(f"Variancia: {idade.var():.2f}")
print(f"Desvio Padrao: {idade.std():.2f}")
print(f"Amplitude: {(idade.max() - idade.min()):.2f}")
print(f"Q1 (25%): {idade.quantile(0.25):.2f}")
print(f"Q2 (50%): {idade.quantile(0.50):.2f}")
print(f"Q3 (75%): {idade.quantile(0.75):.2f}")

print("\n=== VARIAVEL 2: TARIFA ===")
print(f"Media: {df['Fare'].mean():.2f}")
print(f"Mediana: {df['Fare'].median():.2f}")
print(f"Moda: {df['Fare'].mode()[0]:.2f}")
print(f"Variancia: {df['Fare'].var():.2f}")
print(f"Desvio Padrao: {df['Fare'].std():.2f}")
print(f"Amplitude: {(df['Fare'].max() - df['Fare'].min()):.2f}")
print(f"Q1 (25%): {df['Fare'].quantile(0.25):.2f}")
print(f"Q2 (50%): {df['Fare'].quantile(0.50):.2f}")
print(f"Q3 (75%): {df['Fare'].quantile(0.75):.2f}")
```
